```python
class CodeWord(object):
    '''
    Cada codeword contem um v = (R, G, B) tal como um 
    aux = (I_min, I_max, f, lambda, p, q), tal como definido
    na imagem abaixo
    '''
    
    def __init__(self, v, aux):
        self.v = v
        self.aux = aux
        
        
class Aux(object):
    ''' Vide imagem abaixo '''
    
    def __init__(self, i_min, i_max, f, l, p, q):
        self.I_max = i_max
        self.I_min = i_min
        self.f = f
        self.l = l  # lambda e uma palavra reservada :(
        self.p = p
        self.q = q
```

```python
from math import sqrt, inf

def colordist(x, v):
    ''' implementa o algoritmo acima '''
    xt2 = sum([xx**2 for xx in x])
    vi2 = sum([vv**2 for vv in v])
    xv2 = sum([xx * vv for xx, vv in zip(x, v)])**2
    try:
        p2 = xv2 / vi2
    except:
        p2 = 0
    try:
        res = sqrt(xt2 - p2)
    except ValueError as e:
        res = inf
    return res


def brightness(x, i_best, alpha, beta):
    ''' true if i_low <= x <= i_hi '''
    i_min, i_max = i_best
    i_low = alpha * i_max
    i_hi = min((beta * i_max), (i_min / alpha))
    return i_low <= x <= i_hi


def update_codebook(pixel, words, t, e_1, alpha, beta):
    ''' implementa o algoritmo da imagem acima '''
    # (I)
    # (II)
    # para cada pixel ao longo do tempo...
    # (i)
    pixel = [float(p) for p in pixel]
    x_t = pixel
    I = sqrt(sum([k**2 for k in pixel]))
    # (ii)
    matched_cw = None
    if len(words) != 0:
        for word in words:
            v_m = word.v
            if (colordist(x_t, v_m) <= e_1 and
                brightness(I, (word.aux.I_min,
                               word.aux.I_max), 
                           alpha, beta)):
                matched_cw = word
                break
    # (iii) 
    if len(words) == 0 or matched_cw is None:
        words.append(CodeWord(x_t, Aux(I, I, 1, t-1, t, t)))
        v_m = x_t
    # (iv)
    else:
        v_m = matched_cw.v
        aux_m = matched_cw.aux
        for i in range(3):
            matched_cw.v[i] = ((aux_m.f * v_m[i] + pixel[i]) /
                        (aux_m.f + 1))
        matched_cw.aux.I_max = max(I, aux_m.I_max)
        matched_cw.aux.I_min = min(I, aux_m.I_min)
        matched_cw.aux.f += 1
        matched_cw.aux.l = max(aux_m.l, t-aux_m.q)
        matched_cw.aux.q = t

    return words
        
                
def generate_codebooks(imgs, shape, e_1=12, alpha=0.6, beta=1.1):
    T = len(imgs)
    
    codebooks = np.empty((shape[0], shape[1]), dtype=object)
    for i in range(shape[0]):
        for j in range(shape[1]):
            codebooks[i][j] = []
    
    print(shape)
    
    for t in range(T):
        img = imgs[t]
        for i in range(shape[0]):
            for j in range(shape[1]):
                # pixel i,j ao longo do tempo
                codebooks[i][j] = update_codebook(img[i][j], 
                                               codebooks[i][j],
                                               t, e_1, alpha,
                                               beta)
        print(f'done with img[{t}]')
        
    for i in range(shape[0]):
        for j in range(shape[1]):
            N = len(codebooks[i][j])
            for word in codebooks[i][j]:
                word.aux.l = max(word.aux.l, 
                             (N - word.aux.q + word.aux.p - 1))
    np.save('test_save.npy', codebooks)
    return codebooks
```

```python
FG = (255, 255, 255)
BG = (0, 0, 0)

def extract_bg(img, codebooks, e2=12, alpha=0.4, beta=1.1):
    '''
    codebooks is a matrix of the same 
    dims as the original img 
    '''
    shape = codebooks.shape
    bgs = img.copy()
    sumb = sumd = 0
    for i in range(shape[0]):
        for j in range(shape[1]):
            cws = codebooks[i][j]
            pixel = img[i][j]
            pixel = [float(p) for p in pixel]
            I = sqrt(sum([k**2 for k in pixel]))
            matched = False
            for cw in cws:
                bright = brightness(I, 
                                    (cw.aux.I_min, 
                                     cw.aux.I_max), 
                                    alpha, beta)
                dist = colordist(pixel, cw.v) <= e2
                if bright:
                    sumb += 1
                if dist:
                    sumd += 1
                if dist and bright:
                    matched = True
                    break
            bgs[i][j] = BG if matched else FG
        if i % 100 == 0:
            print(f'done with line {i}')
    print(f'sum bright: {sumb}')
    print(f'sum dist: {sumd}')
    return bgs
            
```
